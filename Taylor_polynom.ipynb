Polinomios de taylor: Evaluaciones y gráficas.

Aproximación de $arctan(0.25)$ usando el polinómio de Taylor de $arctan(x)$ alrededor de $x_0=0$ y error relativo: \
\
$err_n=\frac{T_n-T_{n-1}}{T_n}$ \
\
Donde $(T_n)$ es la sucesión de aproximaciones de $arctan(0.25)$ con el polinómio de taylor de grado $n$.

```python
import pandas as pd

x_0=0
x=0.25

T=0
L_T=[0]
L_err=[1]
i=0
err=1
while abs(err)>0.5*10**(-10):
    T=T+((-1)**i)*(x-x_0)**(2*i)
    L_T.append(T)

    err=(L_T[i+1]-L_T[i])/(L_T[i+1])
    L_err.append(err)

    i=i+1

D={"f(-0.25)":L_T,
    "Error":L_err}
df=pd.DataFrame(D)

print(df.iloc[1:i+1][:])
```

Gráfica de los polinómios de Taylor de $ln$ alrededor de $1$ con $x \in (0,2)$

```python
import numpy as np
import matplotlib.pyplot as plt

X=np.linspace(0,2,100)
Ln=np.log(X)

T=np.array([0])
for n in range(1,17):
    T=T+((-1)**(n+1)*(X-1)**n)/n
    #equivalente a pedir que n=2^m
    if (n&(n-1))==0:
        plt.plot(X,T,label=f"Grado {n}")

plt.plot(X,Ln,label="ln(x)",linewidth=3)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("ln(x) Taylor series")
plt.legend(
    loc='best',
    fontsize=10,
    frameon=True,
    title='Curvas'
)
plt.grid()

plt.show()
```
