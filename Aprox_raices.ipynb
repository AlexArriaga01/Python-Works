Métodos de aproximación de raices en una variable.

Método de la bisección.

Fuerza bruta, usa el teorema del valor intermedio para ir cazando la raíz bisecando el intervalo inicial.

```python
import numpy as np
import pandas as pd

def f(x):
    return x**2-2*np.cos(x)-x+1

a,b=0,2

def bi(a,b):
    if f(a)*f(b)<0:
        p=(a+b)/2
        if f(a)*f(p)<0:
            a,b=a,p
        elif f(p)*f(b)<0:
            a,b=p,b
    else:
        raise ValueError("Intervalo inválido")
    return a,b,p

A=[a]
B=[b]
P=[0]
Fp_n=[f(0)]
for i in range(25):
    a,b,p=bi(a,b)
    A.append(a)
    B.append(b)
    P.append(p)
    Fp_n.append(f(p))

D={
    "a_n":A,
    "b_n":B,
    "p_n":P,
    "f(p_n)":Fp_n
}
DF=pd.DataFrame(D)

print(DF)
```

Método del punto fijo. \
\
Se debe encontrar una expresión de la forma $x=g(x)$ para $f(x)=0$ donde $g$ es contractiva i,e, satisface el teorema del punto fijo de Banach. Una condición suficiente para que el punto fijo sea único es que $|g'(x)| \leq k<1$ en una vecindad cerrada del punto fijo.

```python
def f(x):
  return x**3+4*x**2-10

def g(x):
  return (1/2)*((-x**3+10)**(1/2))

p=1.5
print("x \t \t \t f(x)")
for i in range(25):
  p=g(p)
  print(p,"\t",f(p))
```

Método de Newton. \
\
Si $r$ es raíz de $f$ y $f'(r) \neq 0$ entonces alrededor de $r$, se puede iterar el método de Newton para un punto inicial $p_0$ y este convergerá a $r$. Note que este método es sensible a las condiciones iniciales.

```python
import numpy as np
import pandas as pd

def f(x):
  return np.cos(x)-3*np.sin(x)-x**3

def df(x):
  return -np.sin(x)-3*np.cos(x)-3*x**2

p=0.5
R=[p]
E=[p]
for i in range(10):
  p=p-(f(p)/df(p))
  R.append(p)
  E.append(f(p))
  
D={
    "Raíz":R,
    "Evaluación":E
}
print(pd.DataFrame(D))
```

Método de la secante. \
\
Cuando $f'(r)=0$, se puede usar una recta secante a la curva, en vez de la recta tangente de la derivada. Converge muy rápido igual que el método de Newton.

```python
import numpy as np
import pandas as pd

def f(x):
    return np.sin(x)-x**2+3*x-1

pm_1,pm=0,1

P=[pm_1,pm]
fP=[f(pm_1),f(pm)]
while abs(f(pm))>0.5*10**(-10):
    p=pm
    pm=pm_1-(((pm_1-pm)*f(pm_1))/(f(pm_1)-f(pm)))
    pm_1=p

    P.append(pm)
    fP.append(f(pm))

DF=pd.DataFrame({"p_m":P,"f(p_m)":fP})
print(DF)
```

Método de la posición falsa (regula falsi). \
\
Es algo parecido a la bisección, solo que en vez de calcular el punto medio de $[a,b]$, calcula la intersección de la recta que une $f(a)$ con $f(b)$. Este método es robusto (casi siempre converge), pero no converge tan rápido.

```python
import numpy as np
import pandas as pd

def f(x):
    return x**3-3*x+1

a,b=0,1

def bi(a,b):
    if f(a)*f(b)<0:
        p=(a*f(b)-b*f(a))/(f(b)-f(a))
        if f(a)*f(p)<0:
            a,b=a,p
        elif f(p)*f(b)<0:
            a,b=p,b
    else:
        raise ValueError("Intervalo inválido")
    return a,b,p

A=[a]
B=[b]
P=[0]
Fp_n=[f(0)]

for i in range(10):
    a,b,p=bi(a,b)
    A.append(a)
    B.append(b)
    P.append(p)
    Fp_n.append(f(p))

D={
    "a_n":A,
    "b_n":B,
    "p_n":P,
    "f(p_n)":Fp_n
}

print(pd.DataFrame(D))
```
