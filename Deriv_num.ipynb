Métodos de derivación numérica

El método de diferencias centradas para una función $f$ tiene un error de $O(h^2)$ donde $h=x-x_0$. Se deriva de la expansión de Taylor de $f$ alrededor de $x_0$. \
\
El método de diferencias hacia atrás tiene error $O(h)$. \
\
$h$ debe ser suficientemente pequeño para mejor aproximación, pero no muy pequeño para evitar que se cancelen cosas importantes! \
\
El método de Richardson combina dos aproximaciones con un método pero con diferente $h$ para hacer una mejor aproximación algebraicamente. Donde $p$ es el orden del error del método. \
\
Ejemplo con la función: \
\
$f(x)=sin(cos(10x^2))-0.5x+1$

```python
import numpy as np

def f(x):
    return np.sin(np.cos(10*x**2))-0.5*x+1

def DifCen(f,x,h):
    #p=2, i,e error= O(h^2)
    t_i=x
    ti=[t_i,t_i+h,t_i+2*h,t_i-2*h,t_i-h]
    return (1/(12*h))*(-f(ti[2])+8*f(ti[1])-8*f(ti[-1])+f(ti[-2]))

def DifAtr(f,x,h):
    #p=1, i,e error= O(h^1)
    t_i=x
    ti=[t_i,t_i+h,t_i+2*h]
    return (4*f(ti[1])-3*f(ti[0])-f(ti[2]))/(2*h)

def Rich(f,x,h_1,h_2,df,p):
    D_1=df(f,x,h_1)
    D_2=df(f,x,h_2)
    return (D_2*h_1**p-D_1*h_2**p)/(h_1**p-h_2**p)

print(f"Diferencias Centradas: f'(1)={DifCen(f,1,0.001)}")
print(f"Diferencias hacia atrás: f'(1)={DifAtr(f,1,0.001)}")
print(f"Richardson con DifCen: f'(1)={Rich(f,1,0.001,0.00005,DifCen,2)}")
```
