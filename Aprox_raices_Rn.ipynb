Métodos de aproximación de soluciones para sistemas de ecuaciones. \
\
Un sistema de ecuaciones de $n$ ecuaciones y $n$ variables se puede ver como una aplicación $F:R^n \longrightarrow R^n$, con $F=(f_1,...,f_n)$. Y calcular una "raiz" de $F$ se reduce a resolver $F(X)=0$.

Método del punto fijo. \
\
Se despeja una variable (la que sea más cómoda) de cada una de las ecuaciones y obtener algo de la forma $X=G(X)$: \
\
$x_1=g_1(X)$ \
$...$ \
$x_n=g_n(x_n)$ \
\
Y se aplica la iteración del punto fijo: $X^{n}=G(X^{n-1})$. Esta converge siempre que G sea una contracción. Una condición suficiente para la convergencia es que: $|det(DG(X))| \leq k<1$

```python
import numpy as np
import pandas as pd

def F(X):
    return np.cos(X[1]),np.cos(X[0])

x,y=0,0
X=np.array([x,y])

L_X=[(x,y)]
L_FX=[F(X)]
for i in range(10):
    X=F(X)
    L_X.append(X)
    L_FX.append(F(X))

D={"(x,y)":L_X,
   "F(x,y)":L_FX}

print(pd.DataFrame(D))
```

Método de Newton-Raphson. \
\
Usando el teorema de Taylor, dada la aplicacion F. \
\
$F(X*+H) = F(X*) + DF(X*) H + R(H)$ \
\
Con esto, queremos que $X*$ sea tal que $F(X*)=0$. Si $det(DF(X*)) \neq 0$, entonces existe $DF(X)^{-1}$. Luego, ignoramos el residuo $R$. \
\
Queremos que $F(X^n+ \Delta X)=0$. Usando la fórmula de Taylor otra vez, tenemos que: \
\
$F(X^n)+DF(X^n) \Delta X = 0 \implies \Delta X = - DF(X^n)^{-1} F(X^n)$ \
\
Entonces $X^{n+1}=X^n-\Delta X$. ie, el término $n+1$ es el término $n$ "corregido" con el $\Delta X$ que se encontró. \
\
En la práctica es muy costoso calcular $DF(X)^{-1}$, entonces se usa descomposición $LU$ para resolver el sistema lineal $F(X^n)+DF(X^n) \Delta X = 0$. Esto lo hace el comando: np.linalg.solve($DF(X^n),F(X^n)$)

```python
import numpy as np
import pandas as pd

def F(X):
    return np.array([np.cos(X[1])-X[0],np.cos(X[0])-X[1]])

def DF(X):
    return np.array([[-1,-np.sin(X[1])],[-np.sin(X[0]),-1]])

x,y=0,0
X=np.array([x,y])

L_X=[X]
L_FX=[F(X)]
for i in range(10):
    delta=np.linalg.solve(DF(X),F(X))
    X=X-delta
    L_X.append(X)
    L_FX.append(F(X))

D={"(x,y)":L_X,
   "F(x,y)":L_FX}

print(pd.DataFrame(D))
```
